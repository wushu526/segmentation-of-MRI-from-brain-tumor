```python
import csv

filename = '/Users/wushu/Desktop/MSc Project/Project/UNet2D/models/ShuWu_Unet_woDS/log.csv'
with open(filename) as f:
        reader = csv.reader(f)
        header_row = next(reader)
        
        for index, column_header in enumerate(header_row):
            print(index, column_header)
        epochs = []
        losss = []
        val_losss = []
        ious = []
        val_ious = []
        for row in reader:
            epoch = float(row[0])
            epochs.append(epoch)
            
            loss = float(row[2])
            losss.append(loss)
            
            val_loss = float(row[4])
            val_losss.append(val_loss)
            
            iou = float(row[3])
            ious.append(iou)
            
            val_iou = float(row[5])
            val_ious.append(val_iou)
        
        print(epochs)
        print(losss)
        print(val_losss)
```

```python
from matplotlib import pyplot as plt
```

```python
x_data = epochs
y_data_1 = losss
y_data_2 = val_losss

ln1, = plt.plot(x_data,y_data_1,color='red')
ln2, = plt.plot(x_data,y_data_2,color='blue')

plt.xlabel('epochs')
plt.grid(axis="y")
plt.legend(handles=[ln1,ln2],labels=['loss','val_loss'])
plt.show()
```

```python
x_data = epochs
y_data_1 = ious
y_data_2 = val_ious

ln1, = plt.plot(x_data,y_data_1,color='red')
ln2, = plt.plot(x_data,y_data_2,color='blue')
plt.xlabel('epochs')
plt.grid(axis="y")
plt.legend(handles=[ln1,ln2],labels=['iou','val_iou'])
plt.show()
```
